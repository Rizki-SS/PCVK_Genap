```
# 1. Tampilkan image dalam channel Red-Green dan Green-Blue saja!

img_rg = img.copy()
img_rg [:,:,2] = 0    #Blue channel di set 0

img_gb = img.copy()
img_gb [:,:,0] = 0    #Red channel di set 0
imgConcat = cv.hconcat((img_rg,img_gb))

fig = plt.figure(figsize=(15,15))
ax = fig.add_subplot(111)
plt.imshow(imgConcat)
```

```
#2. Tampilkan image baris ke 10-100, kolom 10-100!

imgTgs2 = img[10:100, 10:100]
imgCrop = cv.cvtColor(imgTgs2, cv.COLOR_BGR2RGB)
plt.imshow(imgCrop)
```

```
# 3. Tampilkan image baris ke 5-30, semua kolom, channel Green saja!
img_sc2 = img[5:30, :]
img_sc2 [:, :, 0] = 0
img_sc2 [:, :, 2] = 0
plt.imshow(img_sc2)
```

```
# 4. Buat 5 kotak berbagai ukuran dan warna yang berbeda dalam satu image.
#    disarankan menggunakan bilangan acak/random!

import random
canvas=np.zeros(shape=(800,800,3),dtype=np.int32)

for i in range(6):
  k1 = random.randint(0,799)
  k2 = random.randint(0,799)
  kotak = cv.rectangle(canvas, pt1=(k1, k1), pt2=(k2, k2), color=(k1, k2, random.randint(0,255)),thickness=5)
  plt.imshow(kotak)
```

```
# 5. Tampilkan image dengan posisi terbalik!

img6 = cv.imread('/content/drive/MyDrive/Colab Notebooks/503.jpg')
img6 = cv.flip(cv.cvtColor(img6,cv.COLOR_BGR2RGB),-1)
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111)
ax.imshow(img6)
```

```
# 6. Buat rectangle dan circle pada bagian wajah dari image kitten!

img_sc7 = cv.cvtColor(img, cv.COLOR_BGR2RGB)
cv.rectangle(img_sc7,pt1=(100,450),pt2=(550,50),color=(255,255,255),thickness=3)
cv.circle(img_sc7,center=(350,250),radius=200,color=(255,0,0),thickness=3)
plt.imshow(img_sc7)
```
